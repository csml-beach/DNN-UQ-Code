```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


# Load the CSV file into a DataFrame
file_path = "results_bnn.csv"  # Replace with the path to your CSV file
df = pd.read_csv(file_path)

# Calculate means and standard deviations of every column
column_means_bnn = df.mean()
column_std_devs_bnn = df.std()

# Print the results
print("Means:")
print(column_means_bnn)

print("\nStandard Deviations:")
print(column_std_devs_bnn)
```
